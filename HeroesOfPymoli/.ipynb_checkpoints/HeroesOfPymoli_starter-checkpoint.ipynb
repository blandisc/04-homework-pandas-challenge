## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins =[0, 9, 14, 19, 24, 29, 34, 39, 99]

labels = ["<10","10-14","15-19", "20-24", "25-29", "30-34", "35-39",
         "40+"]

purchase_data_df_nd["Age Range"] = pd.cut(
    purchase_data_df_nd["Age"],
    bins,
    labels=labels)


summary_demo_df = pd.DataFrame({
    
    "Total Count" : purchase_data_df_nd.groupby(
        "Age Range")["Age"].count(),
    "Percentage of Players": (summary_demo_df["Total Count"] / playerCount) * 100
    
})
#Formating
th_props = [
  ('font-size', '14px'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', '#F7F5FB'),
  ('background-color', '#324A5F')
  ]

td_props = [
  ('font-size', '13px'),
    ('text-align', 'center')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]


(summary_demo_df.style
    .format({'Percentage of Players': "{:.2f}%"})
    .set_table_styles(styles))
```
