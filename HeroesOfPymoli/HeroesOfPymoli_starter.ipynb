```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#Eliminating duplicates
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df_nd = purchase_data.drop_duplicates(subset = "SN")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Introduciong data into a df

purchasingAnalysis_df = pd.DataFrame({
    "Number of Unique Items" : purchase_data["Item Name"].nunique(),
    "Average Price" : [purchase_data["Price"].mean()],
    "Number of Purchases" : purchase_data["Purchase ID"].nunique(),
    "Total Revenue" : [purchase_data["Price"].sum()]
})


#Formating
th_props = [
  ('font-size', '14px'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', '#F7F5FB'),
  ('background-color', '#324A5F')
  ]

td_props = [
  ('font-size', '13px'),
    ('text-align', 'center')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]


(purchasingAnalysis_df.style
    .format({'Average Price': "${:.2f}"})
     .format({'Total Revenue': "${:.2f}"})
    .set_table_styles(styles))
```

## Gender Demographics


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Creating DF (Maintained MFO order for consistency)
purchasing_analysis_g_df = pd.DataFrame ({
    "Gender" : ["Male", "Female", "Other / Non-Disclosed"],
    "Purchase Count" : purchase_data_df.groupby("Gender"
                                               )["Purchase ID"].count(),
    "Average Purchase Price" : purchase_data_df.groupby("Gender"
                                               )["Price"].mean(),
    "Total Purchase Value" : purchase_data_df.groupby("Gender"
                                               )["Price"].sum(),
    "Avg Total Purchase per Person" : purchase_data_df.groupby("Gender"
                                               )["Price"].sum()/
    purchase_data_df_nd["Gender"].value_counts()
    
})

purchasing_analysis_g_df = purchasing_analysis_g_df.set_index("Gender")

#Formating
th_props = [
  ('font-size', '14px'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', '#F7F5FB'),
  ('background-color', '#324A5F')
  ]

td_props = [
  ('font-size', '13px'),
    ('text-align', 'center')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]


(purchasing_analysis_g_df.style
    .format({'Average Purchase Price': "${:.2f}"})
    .format({'Total Purchase Value': "${:,.2f}"})
    .format({'Avg Total Purchase per Person': "${:,.2f}"})
    .set_table_styles(styles))
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins =[0, 9, 14, 19, 24, 29, 34, 39, 99]

labels = ["<10","10-14","15-19", "20-24", "25-29", "30-34", "35-39",
         "40+"]
# I introduced an extra columns to see the composition of each age range per gender
maleF = purchase_data_df["Gender"] == "Male"
femaleF = purchase_data_df["Gender"] == "Female"

purchase_data_df_nd["Age Ranges"] = pd.cut(
    purchase_data_df_nd["Age"],
    bins,
    labels=labels)

purchase_data_male_nd = purchase_data_df_nd.loc[maleF]
purchase_data_female_nd = purchase_data_df_nd.loc[femaleF]
otherF = purchase_data_df["Gender"] == "Other / Non-Disclosed"

summary_demo_df = pd.DataFrame({
    
    "Total Count" : purchase_data_df_nd.groupby(
        "Age Ranges")["Age"].count(),
    "Percentage of Players": purchase_data_df_nd.groupby(
        "Age Ranges")["Age"].count() / playerCount * 100,
# I introduced an extra columns to see the composition of each age range per gender
    "Males" : purchase_data_male_nd.groupby(
        "Age Ranges")["Age"].count(),
    "Females" : purchase_data_female_nd.groupby(
        "Age Ranges")["Age"].count()
    
})
#Formating
th_props = [
  ('font-size', '14px'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', '#F7F5FB'),
  ('background-color', '#324A5F')
  ]

td_props = [
  ('font-size', '13px'),
    ('text-align', 'center')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]


(summary_demo_df.style
    .format({'Percentage of Players': "{:.2f}%"})
    .set_table_styles(styles))
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins =[0, 9, 14, 19, 24, 29, 34, 39, 99]

labels = ["<10","10-14","15-19", "20-24", "25-29", "30-34", "35-39",
         "40+"]

purchase_data_df["Age Ranges"] = pd.cut(
    purchase_data_df["Age"],
    bins,
    labels=labels)

summary_ages_df = pd.DataFrame({
    
    "Purchase Count": purchase_data_df.groupby(
    "Age Ranges")["Purchase ID"].count(),
    
    "Average Purchase Price": purchase_data_df.groupby(
    "Age Ranges")["Price"].mean(),
                                
    "Total Purchase Value" : purchase_data_df.groupby(
    "Age Ranges")["Price"].sum(),
    
    "Avg Total Purchase per Person" : purchase_data_df.groupby(
    "Age Ranges")["Price"].sum()/ purchase_data_df_nd.groupby(
        "Age Ranges")["Age"].count()
     
})

#Formating
th_props = [
  ('font-size', '14px'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', '#F7F5FB'),
  ('background-color', '#324A5F')
  ]

td_props = [
  ('font-size', '13px'),
    ('text-align', 'center')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]


(summary_ages_df.style
    .format({'Average Purchase Price': "${:.2f}"})
    .format({'Total Purchase Value': "${:,.2f}"})
    .format({'Avg Total Purchase per Person': "${:.2f}"})
    .set_table_styles(styles))


```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# purchase_data_df["SN"].value_counts()

top_spenders_df = pd.DataFrame({
    "Purchase Count" : purchase_data_df.groupby("SN")["Purchase ID"].count(),
    "Average Purchase Price" : purchase_data_df.groupby("SN")["Price"].sum()  
    / purchase_data_df.groupby("SN")["Purchase ID"].count(),
    "Total Purchase Value" : purchase_data_df.groupby("SN")["Price"].sum(),
})


top_spenders_df.sort_values(ascending = False, by = "Total Purchase Value").head(5)


#Formating
th_props = [
  ('font-size', '14px'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', '#F7F5FB'),
  ('background-color', '#324A5F')
  ]

td_props = [
  ('font-size', '13px'),
    ('text-align', 'center')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]


(top_spenders_df
    .sort_values(ascending = False, by = "Total Purchase Value")
    .head(5).style
    .format({'Average Purchase Price': "${:.2f}"})
    .format({'Total Purchase Value': "${:.2f}"})
    .set_table_styles(styles))
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame

  

```python
top_items_df = pd.DataFrame({
    
    "Item Name" : purchase_data_df.groupby(
        "Item ID")["Item Name"].first(),
    "Purchase Count" : purchase_data_df.groupby(
        "Item ID")["Purchase ID"].count(),
    "Item Price" : purchase_data_df.groupby(
        "Item ID")["Price"].mean(),
    "Total Purchase Value": purchase_data_df.groupby(
        "Item ID")["Price"].sum(),
    
 })

top_items_df = top_items_df.

#Formating
th_props = [
  ('font-size', '14px'),
  ('text-align', 'center'),
  ('font-weight', 'bold'),
  ('color', '#F7F5FB'),
  ('background-color', '#324A5F')
  ]

td_props = [
  ('font-size', '13px'),
    ('text-align', 'center')
  ]

styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props)
  ]


(top_items_df
    .sort_values(by = "Purchase Count", ascending = False)
    .head(5)
    .style
    .format({'Item Price': "${:.2f}"})
    .format({'Total Purchase Value': "${:.2f}"})
    .set_table_styles(styles))
```
